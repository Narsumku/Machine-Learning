```python
import numpy as np
import pandas as pd

pd.set_option("display.precision", 1)
```

```python
np.random.seed(10120032)

# Mean (Î») of the Poisson distribution
lambda_value = 0.5

# Generate one Poisson-distributed random integer
single_random_int = np.random.poisson(lambda_value)
print(f"Single Poisson random integer: {single_random_int}")

# Generate an array of Poisson-distributed random integers
size = 1000
random_integers = np.random.poisson(lambda_value, size) + 1
print(f"Array of Poisson random integers: {random_integers}")
```

```python
import random

Categories = ['Business', 'Entertainment', 'Politics', 'Sport', 'Tech', 'Healthcare', 'Academic', 'Media & News']
user_cat = [[Categories[item] for item in random.sample(range(8),random_integer)] for random_integer in random_integers]
```

```python
import pandas as pd

speaker_df_ = pd.read_csv(cwd + '/Data/labeled_data.csv')
speaker_df_.insert(0, 'id', [f'speaker_{i}' for i in range (len(speaker_df_))])
```

```python
speaker_df =(speaker_df_[speaker_df_['Category 1'] == 'Politics'].sample(frac = 1, random_state=101).reset_index().iloc[:,1:]).copy()

for i in [0,1,3,4,5,6,7]:
    speaker_df_add = speaker_df_[speaker_df_['Category 1'] == Categories[i]].sample(frac = 1, random_state=101)
    speaker_df = speaker_df._append(speaker_df_add.reset_index().iloc[:26,1:])

speaker_df = speaker_df.reset_index().iloc[:,1:]
speaker_df['id'] = [f'speaker_{i}' for i in range (len(speaker_df))]
```

```python
speaker_df.to_csv(cwd + '/Data/speaker_with_attribute.csv', index=False)
```

```python
random.seed(10)
np.random.seed(10120032)

for j in range (len(user_cat)):
    cat = user_cat[j]
    filtered_speaker = speaker_df[(speaker_df['Category 1'].isin(cat)) |
                                (speaker_df['Category 2'].isin(cat)) |
                                (speaker_df['Category 3'].isin(cat)) ].reset_index().iloc[:,1:]

    get_speaker_to_review = random.sample(range(len(filtered_speaker)), len(filtered_speaker)//5)
    rating_to_speaker_filtered = np.array([np.random.choice([3.5,4,4.5,5]) for i in range (len(get_speaker_to_review))])

    data_i = pd.DataFrame({'user_id':np.array([f'user_{j}' for i in range (len(get_speaker_to_review))]),
                        'speaker_id':filtered_speaker['id'][get_speaker_to_review ].to_numpy(),
                        'rating':rating_to_speaker_filtered,
                        'full name':filtered_speaker['full_name'][get_speaker_to_review ].to_numpy(),
                        'experience':filtered_speaker['Experience'][get_speaker_to_review ].to_numpy(),
                        'category 1':filtered_speaker['Category 1'][get_speaker_to_review ].to_numpy(),
                        'category 2':filtered_speaker['Category 2'][get_speaker_to_review ].to_numpy(),
                        'category 3':filtered_speaker['Category 3'][get_speaker_to_review ].to_numpy()})

    non_filtered_speaker = filtered_speaker = speaker_df[~((speaker_df['Category 1'].isin(cat)) |
                                (speaker_df['Category 2'].isin(cat)) |
                                (speaker_df['Category 3'].isin(cat))) ].reset_index().iloc[:,1:]

    get_non_filtered_speaker_to_review = random.sample(range(len(non_filtered_speaker)), len(non_filtered_speaker) // 50)
    rating_to_speaker_non_filtered = [np.random.choice([2,2.5,3]) for i in range (len(get_non_filtered_speaker_to_review ))]

    data_j = pd.DataFrame({'user_id':np.array([f'user_{j}' for i in range (len(get_non_filtered_speaker_to_review))]),
                        'speaker_id':non_filtered_speaker['id'][get_non_filtered_speaker_to_review ].to_numpy(),
                        'rating':rating_to_speaker_non_filtered,
                        'full name':non_filtered_speaker['full_name'][get_non_filtered_speaker_to_review],
                        'experience':non_filtered_speaker['Experience'][get_non_filtered_speaker_to_review],
                        'category 1':non_filtered_speaker['Category 1'][get_non_filtered_speaker_to_review],
                        'category 2':non_filtered_speaker['Category 2'][get_non_filtered_speaker_to_review],
                        'category 3':non_filtered_speaker['Category 3'][get_non_filtered_speaker_to_review]})

    data_concat = pd.concat((data_i,data_j)).reset_index().iloc[:,1:]

    if j == 0 :
        data = data_concat.copy()
    else:
        data = data._append(data_concat)

data = data.reset_index().iloc[:,1:]
```

```python
data.to_csv(cwd + '/Data/data.csv', index=False)
```

```python
for j in range (size):
    user_data_i = data[data['user_id'] == f'user_{j}']
    num_rating_i = len(user_data_i)
    ave_rating_i = np.mean(user_data_i['rating'])

    ave_rating_per_categories = np.array([0 for i in range (len(Categories))], dtype='float')
    for i,cat in enumerate(Categories):
        mean_val = np.mean(user_data_i[(user_data_i['category 1']).isin([cat]) |
                           (user_data_i['category 2']).isin([cat]) |
                           (user_data_i['category 3']).isin([cat])]['rating'])

        if not np.isnan(mean_val):
            ave_rating_per_categories[i] =  mean_val

    user_side_data_i = pd.DataFrame({
        'user id':[f'user_{j}'],
        'rating count':[num_rating_i],
        'rating ave':[ave_rating_i]
    })

    for i,cat in enumerate(Categories):
        user_side_data_i[cat] = [ave_rating_per_categories[i]]
    if j == 0 :
        user_side_data = user_side_data_i.copy()
    else:
        user_side_data = user_side_data._append(user_side_data_i)

user_side_data = user_side_data.reset_index().iloc[:,1:]
```

```python
for j in range (len(speaker_df)):
    speaker_data_i = data[data['speaker_id'] == f'speaker_{j}']
    ave_rating_i = np.mean(speaker_data_i['rating'])

    ave_val_per_categories = np.array([0 for i in range (len(Categories))], dtype='float')

    for i in range (1,4):
        speaker_cat = speaker_df[speaker_df['id'] == f'speaker_{j}' ][f'Category {i}'].to_numpy()

        if len(speaker_cat) != 0:
            if type(speaker_cat[0]) == str:
                get_idx = Categories.index(speaker_cat[0])
                ave_val_per_categories[get_idx] = 1

    speaker_side_data_i = pd.DataFrame({
        'speaker id':[f'speaker_{j}'],
        'rating ave':[ave_rating_i]
    })

    for i,cat in enumerate(Categories):
        speaker_side_data_i[cat] = [ave_val_per_categories[i]]
    if j == 0 :
        speaker_side_data = speaker_side_data_i.copy()
    else:
        speaker_side_data = speaker_side_data._append(speaker_side_data_i)

speaker_side_data = speaker_side_data.reset_index().iloc[:,1:]
```

```python
user_side_data.to_csv(cwd + '/Data/user_side_data.csv', index=False)
```
